```python
import os
import scipy
import scipy.io
import numpy as np
import matplotlib as plt
%matplotlib inline  
```

Load Data and create set of faces

```python
rawData = scipy.io.loadmat("/home/nathan/QEA-Eigenfaces/data/classdata.mat")['classdata']
```

```python
faces = []

for i in range(rawData.shape[2]):
    faces.append(rawData[::,::,i])
```

Make image displaying functions

```python
def show_image(image):
    return plt.pyplot.imshow(image, cmap='Greys_r')
```
